- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
df["Legendary"].unique()
```

```python
#Assumption: dragon mean(HP) > Grass mean(HP)
# Hypotheses: H0: Dragon mean(HP) <= Grass mean(HP)
# H1: Dragon mean(HP) > Grass mean(HP)
# Statisticql testdf: TWO SAMPLE T-TEST: compare 2 groups means
dragon_hp = df[df['Type 1'] == "Dragon"]["HP"]
grass_hp = df[df['Type 1'] == "Grass"]["HP"]
statistic, p_value = st.ttest_ind(dragon_hp, grass_hp, equal_var = False, alternative = "greater")
if p_value < 0.05:
    print("We reject H0")
else:
    print("fail to reject H0")

```

## H0 is rejected, there is evidence that Dragon mean(HP) is greater than Grass mean(HP).

```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
print(df.shape)
df.head()
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 34)
- Hospital coordinates (-122, 37)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
#Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
# we cosnider a point p1(x1, y1) and point p2(x2, y2): Euclidean: d = sqrt((x2-x1)^2 + (y2-y1)^2)


def dist_to_school(lon, lat):
    """
    dx=(Δlon)×111.32×cos(mean latitude)
    dy=(Δlat)×111.32
    
    1 degree of latitude ≈ 111.32 km
    1 degree of longitude ≈ 111.32 × cos(latitude) km
    """
    lat_school = 34
    lon_school = -118
   
    #calculate the cartisian coordinates of the school and hospital:
    dx = (lon - lon_school) * 111.32 * np.cos (np.radians((lat+lat_school)/2))
    dy = (lat - lat_school) * 111.32
   

    dist_school = np.sqrt(
        dx**2 + dy**2
    )
    return round(dist_school, 2)
    
def dist_to_hsptl(lon, lat):
    lat_hsptl = 37 
    lon_hsptl = -122
    #calculate the cartisian coordinates of the school and hospital:
    dx = (lon - lon_hsptl) * 111.32 * np.cos (np.radians((lat+lat_hsptl)/2))
    dy = (lat - lat_hsptl) * 111.32
   

    dist_hsptl = np.sqrt(
        dx**2 + dy**2
    )
    return round( dist_hsptl , 2)


```

```python
dist_school = dist_to_school(-114.31, 34.19)
dist_hsptl = dist_to_hsptl(-114.31, 34.19)
display(dist_school)
display(dist_hsptl)
```

```python
#Apply the 2 functions on the df
df['dist_from_school_km'] = df.apply(lambda row: dist_to_school(row['longitude'], row['latitude']), axis = 1)
df['dist_from_hospital_km'] = df.apply(lambda row: dist_to_hsptl(row['longitude'], row['latitude']), axis = 1)
df.head()
df.describe()
```

```python
# Divide your dataset into houses close and far from either a hospital or school.
```

```python
df['close_to_facility'] = np.where(
    (df['dist_from_school_km'] < 0.5) |
    (df['dist_from_hospital_km'] < 0.5),
    "close", "far"
)
df_close = df[df['close_to_facility'] == "close"]
df_far = df[df['close_to_facility'] == "far"]


df.head()
```

# Hypothesis
## Null Hypothesis: price_mean near facility <= price_mean far from facility
## H1: price_mean near facility > price_mean far from facility
we use Two-sample independent T-test


```python
from scipy import stats as st
stat, p_value = st.ttest_ind(df_close['median_house_value'], df_far['median_house_value'], alternative = "greater")
if p_value <0.05:
    print(f"Hypothesis rejected: {p_value} < 0.05. There is sufficient evidence that the  price of near houses is greater than far houses")
else:
    print(f"Fail to reject {p_value} >= 0.05.: there is insufficient evidence that the mean price of near houses is greater than far houses")
```
